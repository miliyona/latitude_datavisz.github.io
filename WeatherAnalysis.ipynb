Weather Analysis: Latitude vs Max Temprature, Humidity, Cloudiness, and Wind Speed. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.pyplot import figure
import time
```

```python
cities_df=pd.read_csv(r'Resources\cities.csv')
```

```python
cities_df.rename(columns={'Max Temp':'Max_Temp', 'Wind Speed':'Wind_Speed'}, inplace=True)
cities_df.head()
```

```python
plt.scatter(cities_df.Lat, cities_df.Max_Temp,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8)
plt.title("City Latitude vs. Max Temperature (%s)" % time.strftime("%x"))
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid()
plt.savefig("plots/Fig1.png")
plt.show()
```

```python
plt.scatter(cities_df.Lat, cities_df.Humidity,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8)
plt.title("City Latitude vs. Humidity (%s)" % time.strftime("%x"))
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid()
plt.savefig("plots/Fig2.png")
plt.show()
```

```python
plt.scatter(cities_df.Lat, cities_df.Cloudiness,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8)
plt.title("City Latitude vs. Cloudiness (%s)" % time.strftime("%x"))
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid()
plt.savefig("plots/Fig3.png")
plt.show()
```

```python
plt.scatter(cities_df.Lat, cities_df.Wind_Speed,
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8,)
plt.title("City Latitude vs. Wind Speed (%s)" % time.strftime("%x"))
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid()
plt.savefig("plots/Fig4.png")
plt.show()
```
